# Function defining

```python
"""method for calling index instead of key value"""
def get_percentage_number(feature):
        max_ele_value = feature.value_counts(
        )[feature.value_counts().index[0]]#value count order| index[0] so key and index doesnt clash
        total = len(train.index) #number of rows in train
        p_ab = round(100 * float(max_ele_value)/float(total),2) #return highest occurence element
        return p_ab
```

```python
"""feature to be removed from classification group"""
def remove_feature(feature, list_ref):
    new_feature = [x for x in feature if x not in list_ref]
    return new_feature
```

```python
def remove_high_cardinality(type):
    cardinality_keep = []
    to_drop_features =[]
    for feature in train[type].columns:#iterate through each column name, aka, feature
            if get_percentage_number(train[feature]) > 85.0: #check if highest occurence element > 85%
                to_drop_features.append(feature)
            else:
                cardinality_keep.append(feature)
    return cardinality_keep, to_drop_features
```

### grouping and label encoding nominal features with clear speration of elements for range of saleprice 

#### ms_subclass

using ms_subclass saleprice as reference to group ms_subclass values for test set reference

```python
ms_subclass_high = []
ms_subclass_medium = []
ms_subclass_low = []

for x in train['ms_subclass'].unique():
    y= (train.loc[train['ms_subclass']==x, 'saleprice'].quantile(0.75))
    if (y >= 200_000.0):
        ms_subclass_high.append(x)
    elif ((y >= 140_000.0) & (y < 200_000.0)):
        ms_subclass_medium.append(x)
    else:
        ms_subclass_low.append(x)
        
```

check which elements for into which class

```python
print(ms_subclass_high)
print(ms_subclass_medium)
print(ms_subclass_low)
```

conversion of using ms_subclass saleprice for reclassification

```python
"""conversion of ms_subclass saleprice to map ms_subclass values"""
for x in train['ms_subclass'].unique():
    y= (train.loc[train['ms_subclass']==x, 'saleprice'].quantile(0.75))
    if (y >= 200_000.0):
        train['ms_subclass'] = train['ms_subclass'].replace(x,3)
    elif ((y >= 140_000.0) & (y < 200_000.0)):
        train['ms_subclass'] = train['ms_subclass'].replace(x,2)
    else:
        train['ms_subclass'] = train['ms_subclass'].replace(x,1)
```

```python
train['ms_subclass'].unique()
```

removing ms_subclass nominal and adding into ordinal

```python
nominal = list(set(nominal)-set(['ms_subclass']))
ordinal = ordinal + ['ms_subclass']
```

#### neighborhood

using neighborhood saleprice as reference to group neighborhood values for test set reference

```python
ms_neighborhood_1= []
ms_neighborhood_2= []
ms_neighborhood_3= []
ms_neighborhood_4= []
ms_neighborhood_5= []

"""grouping neighborhood according to their 75th quantile saleprice"""
for x in train['neighborhood'].unique():
    y = train.loc[train['neighborhood']==x,'saleprice'].quantile(0.75)
    if (y > 379040):
        ms_neighborhood_5.append(x)
    elif ((y > 301400.0) & (y <= 379040)):
        ms_neighborhood_4.append(x)
    elif((y > 237842.49999999997)&(y<=301400.0)):
        ms_neighborhood_3.append(x)
    elif((y >159500.0)&(y<=237842.49999999997)):
        ms_neighborhood_2.append(x)
    else:
         ms_neighborhood_1.append(x)
```
