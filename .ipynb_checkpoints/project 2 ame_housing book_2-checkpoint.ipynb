# Function defining

```python
"""method for calling index instead of key value"""
def get_percentage_number(feature):
        max_ele_value = feature.value_counts(
        )[feature.value_counts().index[0]]#value count order| index[0] so key and index doesnt clash
        total = len(train.index) #number of rows in train
        p_ab = round(100 * float(max_ele_value)/float(total),2) #return highest occurence element
        return p_ab
```

```python
"""feature to be removed from classification group"""
def remove_feature(feature, list_ref):
    new_feature = [x for x in feature if x not in list_ref]
    return new_feature
```

```python
"""feature to be added for test classification group"""
def add_feature(feature, list_ref):
    new_feature = [x for x in feature if x in list_ref]
    return new_feature
```

```python
def remove_high_cardinality(type):
    cardinality_keep = []
    to_drop_features =[]
    for feature in train[type].columns:#iterate through each column name, or , feature
            if get_percentage_number(train[feature]) > 85.0: #check if highest occurence element > 85%
                to_drop_features.append(feature)
            else:
                cardinality_keep.append(feature)
    return cardinality_keep, to_drop_features
```

conversion of using ms_subclass saleprice for reclassification

```python
"""conversion of ms_subclass saleprice to map ms_subclass values"""
for x in train['ms_subclass'].unique():
    y= (train.loc[train['ms_subclass']==x, 'saleprice'].quantile(0.75))
    if (y >= 200_000.0):
        train['ms_subclass'] = train['ms_subclass'].replace(x,3)
    elif ((y >= 140_000.0) & (y < 200_000.0)):
        train['ms_subclass'] = train['ms_subclass'].replace(x,2)
    else:
        train['ms_subclass'] = train['ms_subclass'].replace(x,1)
```

# Conclusion and Recommendation

From the initial 82 variables from kaggle, i narrowed it down to the top 50 highest correlated variables after performing data cleaning and related exploratory data analysis. Features that have high cardinality(>85%) from nominal and ordinal features were removed as they do not contribute sufficiently to the saleprices. Features that were proven were strongly correlated; like the various basement squarefeet features were removed, only the total squarefeet remains, to avoid multi-collinearity.<br><br>

Finally, after running the various models on the train-holdout set, lasso regression narrowly edged out ridge regression with a train-test score of 29732.19 to 29716.54, and 32724.8 to 32733.44. Ridge regression used 85 features(from 86 features after one hot encoding) while lasso removed 40 features, and used 46 features. <br><br>

Lasso was then used for fitting and predict the train-test values, and showed that gross living area, neighbourhood,  overall quality of house, building type 4 and roof style 7 are the top 5 factors in driving house prices. Buyers that are looking more affordable houses or require certain features for their house, can look into these coefficients to determine the best match of features to their budget for their ideal house.
